```python
import pandas as pd
import os
import csv

# Define file paths
original_file_path = 'TMDB_movie_dataset_v11.csv'
cleaned_file_path = 'TMDB_movie_dataset_v11_cleaned.csv'
limit_rows = 40000  # Specify the number of rows to limit

# Check if the cleaned file already exists
if not os.path.exists(cleaned_file_path):
    # Copy the content of the original CSV to the cleaned CSV file
    with open(original_file_path, 'r', encoding='utf-8') as infile, open(cleaned_file_path, 'w', encoding='utf-8', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        try:
            row_count = 0
            for row in reader:
                writer.writerow(row)
                row_count += 1
                if row_count >= limit_rows:
                    break
        except csv.Error as e:
            print(f"Error occurred while copying CSV: {e}")

# Load the cleaned CSV file into a DataFrame with the limited rows
try:
    movies = pd.read_csv(cleaned_file_path, nrows=limit_rows)
    print(f"Cleaned CSV file loaded into DataFrame with {limit_rows} rows.")
except pd.errors.ParserError as e:
    print(f"Error occurred while parsing CSV into DataFrame: {e}")

```

```python
movies =movies[['id','title','overview','genres','keywords']]
```

```python
movies.dropna(inplace=True)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']
```

```python
new = movies.drop(columns=['overview','genres','keywords'])
```

```python
new
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=4000,stop_words='english')

vectors = cv.fit_transform(new['tags']).toarray()

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gangster Squad')
```
